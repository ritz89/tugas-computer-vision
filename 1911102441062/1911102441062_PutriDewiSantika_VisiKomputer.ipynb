#### 5. Mencoba konversi warna dari BGR ke RGB

```python
# selanjutnya mengolah gambar 
import matplotlib.pyplot as plt
fig = plt.figure() 
fig.add_subplot(121) # ini memasukkan jumlah baris dan kolom 

img1 = plt.imshow(ikan_hias) 

fig.add_subplot(122) 

ikan_hias2 = cv2.cvtColor(ikan_hias, cv2.COLOR_BGR2RGB) # menggunakan colorRGB yang ditukar warnanya sehingga output warna power ranger tosca-orange berubah menjadi green-blue  

plt.imshow(ikan_hias2) 

plt.show()
```

#### 5. Mencoba konversi warna dari BGR ke GRAY

```python
# selanjutnya mengolah gambar 
import matplotlib.pyplot as plt
fig = plt.figure() 
fig.add_subplot(121) # ini memasukkan jumlah baris dan kolom 

plt.imshow(ikan_hias) 

fig.add_subplot(122) 

ikan_hias2 = cv2.cvtColor(ikan_hias, cv2.COLOR_BGR2GRAY) # menggunakan colorRGB yang ditukar warnanya sehingga output warna power ranger tosca-orange berubah menjadi green-blue 

plt.imshow(ikan_hias2) 

plt.show()
```

#### 6. Menambahkan intensitas brightness pada gambar ikan hias menggunakan fungsi **addWeighted()**

```python
import numpy as np
ikan_hias = cv2.imread('ikan-hias-air-tawar-aquarium-6-new.jpg')

cv2_imshow(ikan_hias) 

intensitas_contrast = 1 
intensitas_brightness = 60
result = cv2.addWeighted(ikan_hias, intensitas_contrast, np.zeros(ikan_hias.shape, ikan_hias.dtype), 0, intensitas_brightness) 

cv2_imshow(result)
```
